# Steps:

### 1. Scale Data
=> To ensure all features are treated equally, there are no dominant dimensions in the clustering process. eg: value_eur has a much greater value compared to other features

### 2. Initialize random centroids
=> Pick centroids at random before arriving at updated centroids and the point of convergence.

### 3. Label data points
=> Based on the distance between each data point and initially assigned centroids. Providing cluster assignment for each football player 

### 4. Update centroids 
=> Find center point / mean of each cluster and update centroids.

### 5. Iterate steps 3 and 4 until centroids stop changing (Converge)
=> Repeating the steps of labelling data points/ assigning them to clusters and updating centroids accordingly until the centroids no longer update or change.

## I. Scaling the Data
Min-Max Scaling implemented. 
- Each column in dataframe is first subtracted with minimum values of each column in dataframe to set minimum value to 0.
- Dividing numerator value from above step by result of subtraction of max and min values of column in dataframe to rescale from 0 to 1.
- Multiply the result by 9 and then add 1 to rescale everything on the scale of 1 to 10.

```python
dataFrame = ((dataFrame - dataFrame.min())/(dataFrame.max()-dataFrame.min())) * 9 + 1
```

## II. Initialize random centroids
Iterate through each column/feature through apply() method on dataframe 'data' and select a single,random value using sample() method. 

Each centroid created by randomly sampling values for each feature is appended to a list of centroids in function initial_centroids which takes dataframe(data) and number of clusters(k) as parameters.

Convert the value from panda series to number using float.

The function returns dataframe of each centroid panda series.

```python
def initial_centroids(dataFrame, k):
    centroids = []
    # Number of centroids = number of clusters (k)
    for i in range(k):
        centroid = dataFrame.apply(lambda x: float(x.sample())) 
        
        # Each centroid created by randomly sampling values for each column/feature is appended to list of centroids 
        centroids.append(centroid)
    # Combines each centroid panda series into a dataframe
    return pd.concat(centroids,axis=1)
```

```python
# Calling function on 'data' to generate 5 clusters (k=5)
centroids = initial_centroids(dataFrame,5)
centroids
```

Here, each column is a centroid and each row is a feature.

## III. Label data points
In this step, the euclidean distance between each data point in the dataframe and the cluster centroid is calculated.

Then, the cluster centroid that is closest to each data point is identified.

In function fetch_labels, 
- Applying euclidean distance formula for each data point and all centroids features and storing the result in 'distances' variable.
- returning index/cluster with least distance between data point and centroid of the cluster.

```python
# Calculating distance between each data point and cluster centroid
eucl_distances = centroids.apply(lambda x:np.sqrt(((dataFrame - x)**2).sum(axis=1)))
eucl_distances
```

```python
# To identify indexes (clusters) with least distance from their centroid value
eucl_distances.idxmin(axis=1)
```

```python
def fetch_labels(dataFrame,centroids): 
    # Calculating distance between each data point and cluster centroid
    eucl_distances = centroids.apply(lambda x:np.sqrt(((dataFrame - x)**2).sum(axis=1)))
    
    # To identify indexes (clusters) with least distance from their centroid values
    return eucl_distances.idxmin(axis=1)
```

```python
# Function get_labels output
labels = fetch_labels(dataFrame,centroids)
labels
```

```python
# To display count(number of data points) assigned to each cluster
label_count = labels.value_counts()
label_count
```

## IV. Update Centroids
In this step, geometric mean is calculated to find the updated/new centroid for a cluster in a euclidean space. 

Geometric mean is calculated by multiplying each data point together and finding its n^th root based on the number of data points in each cluster

In the code below, the dataframe is grouped by labels.Then, the log of all feature values for each data point is calculated, followed by its mean. Finally its exponent value is calculated to yield the geometric mean.

This is applied for each label(cluster).

```python
def updated_centroids(dataFrame,labels,k):
    # Grouping dataframe by labels and applying geometric mean, then transposing the dataframe
    return dataFrame.groupby(labels).apply(lambda x: np.exp(np.log(x).mean())).T
```

Dataframe is transposed to obtain features as rows and clusters/labels as columns. The values highlight the geometric mean for each feature for all clusters.

## V. Repeat Steps III and IV until centroids stop changing / Plotting the graph for each iteration

```python
# PCA (Principal Component Analysis) for data visualization (For 5D or features to 2D visualization)
from sklearn.decomposition import PCA

# For plotting
import matplotlib.pyplot as plt

# To Clear Jupyter notebook output and plot new visualization output for each iteration
from IPython.display import clear_output
```

Initializing PCA model and transforming dataframe with 5 features into data with 2 columns (2D data).

```python
def clusterPlots(dataFrame,labels,centroids,iterator):
    
    # Set compress the number of components/columns to display to 2 (from 5 features)
    pca = PCA(n_components=2)
    
    # PCA transform dataframe and fit it
    data_2d = pca.fit_transform(dataFrame)
    
    # PCA transform centroids and transpose accordingly
    centroids_2d = pca.transform(centroids.T)
    
    # Clear output on Jupyter notebook 
    clear_output(wait=True)
    
    # To display each iteration on plotting graph
    plt.title(f'Iteration {iterator}')
    
    # To set x-axis and y-axis of plot and color each label/cluster in the graph for all iterations 
    plt.scatter(x=data_2d[:,0], y=data_2d[:,1],c=labels)
    
    # To set x-axis and y-axis of plot for all iterations 
    plt.scatter(x=centroids_2d[:,0],y=centroids_2d[:,1])
    
    # To display plot
    plt.show()
```

## Implementing Elbow Method to find optimal value of K (Number of Clusters)
- Imported KMeans from sklearn.cluster module to fit data into model. 
- PCA (Principal Component Analysis) has been implemented for dimensionality reduction to reduce number of components/features from 5 to 2 before fitting the data into model.
- Appended inertia values for each value of k (number of clusters) into 'distortions' list
- List values and k values implemented to Elbow Plot.

```python
from sklearn.cluster import KMeans

# Compress and set the number of components/columns to display to 2 (from 5 features)
pca = PCA(n_components=2)
    
# PCA transform dataframe and fit it
data_2d = pca.fit_transform(dataFrame)

# List to append inertia values for each KMeans model with k clusters
distortions = []
K = range(1,10)

# For loop to iterate through number of clusters k
for k in K:
    model = KMeans(n_clusters=k)
    # Fitting PCA transformed data into KMeans model
    model.fit(data_2d)
    
    # Appending inertia values of model to list
    distortions.append(model.inertia_)

# Plotting Elbow method to identify optimal value of K
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

It seems that k=3 is the optimal value. So we'll consider 3 as the optimal number of clusters.

```python
# Number of clusters (optimal value obtained from Elbow plot)
k = 3

# Maximum iterations to convergence of centroids
max_iterations = 100

#Implementing random_centroids function to randomly initialize clusters
centroids = initial_centroids(data, k)

prev_centroids = pd.DataFrame()

# Initial iterator variable
iterator = 1

# Using While loop to iterate 
# To stop loop upon reaching maximum iterations or old centroid is equal to updated centroid
while iterator < max_iterations and not centroids.equals(prev_centroids):
    # Set old centroid to updated centroid for each iteration
    prev_centroids = centroids
    
    # Assign datapoints to its respective clusters with centroids
    labels = fetch_labels(dataFrame, centroids)
    
    #Update centroids based on geometric mean
    centroids = updated_centroids(dataFrame,labels,k)
    
    # Plot clusters
    clusterPlots(dataFrame,labels,centroids,iterator)
    
    # Increment iterator
    iterator += 1
```
