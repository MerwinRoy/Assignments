# Question:

Q.Please write below the *simplest possible* implementation of Fibonacci numbers, given what we learned above. Those are the series where the next element is the sum of the previous two. Also, give me the first 100 Fibonacci numbers.

# Solution:

Ans: 
- I have used a function named fibonacci() that takes an argument which indicates the number upto which fibonacci sequence is created.

- The first two numbers of sequence are placed in variables 'n1' and 'n2' and they are added in a while loop to create sequence upto 'nth' number.

- In every iteration of the loop, the numbers are added and stored in variable 'res' after which the 'n2' variable value is swapped to 'n1' variable and 'res' variable value is swapped to 'n2' variable. 

- There is also a counter variable 'count' which increments after every iteration till the number 'n' passed as argument in function. 

This implementation of producing a fibonacci sequence is based on the hint given to us in one of the ipynb files - 'Introduction-to-Python-dino-part2'

```python
def fibonacci(n):
    count = 0
    n1 = 0
    n2 = 1
    if n <0:
        print("Please enter a positive number.")
    elif n==0 or n==1:
        print("Fibonacci Sequence upto {} terms: {}".format(n,n1))
    else:
        print("Fibonacci sequence: ")
        while count < n:
            print(n1)
            res = n1 + n2
            n1 = n2
            n2 = res
            count+=1
```

Fibonacci Sequence of first 100 terms:

```python
fibonacci(100)
```
