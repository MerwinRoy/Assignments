### Function to read data from csv file

```python
import csv
def read_csv(filename):
    dataset = []
    with open(filename) as f:
        reader = csv.reader(f)
        for i in reader:
            dataset.append(i)
    return dataset
```

```python
dataset = read_csv('pima-indians-diabetes.data.csv')
dataset
```

### Converting datatype in csv file from String to Float

```python
def convert_to_float(dataset):
    for i in range(len(dataset)):
        for j in range(len(dataset[i])):
            dataset[i][j] = float(dataset[i][j])
    return dataset
```

```python
dataset = convert_to_float(dataset)
dataset
```

### Scaling data using minmax scaling

```python
def min_max_range(dataset):
    mmr_list = []
    for i in range(len(dataset[0])):
        col_val = [j[i] for j in dataset]
        min_ = min(col_val)
        max_ = max(col_val)
        mmr_list.append([min_,max_])
    return mmr_list
```

```python
minmax = min_max_range(dataset)
minmax
```

### Normalizing data in the csv dataset file using normalization function

Dataset and minmax value are taken as parameters to scale the data based on minimum and maximum values fetched from minmax function above.

```python
def normalization(dataset,minmax):
    for i in range(len(dataset)):
        for j in range(len(dataset[0])):
            n = dataset[i][j] - minmax[j][0]
            d = minmax[j][1] - minmax[j][0]
            dataset[i][j] = n/d
    return dataset
```

Dataset values scaled in the range of 0 - 1

```python
dataset = normalization(dataset,minmax)
dataset
```

Splitting dataset into training and testing data.

```python
from random import shuffle
def train_test_split(dataset):
    shuffle(dataset)
    n = int(0.8*len(dataset))
    
    train_data = dataset[:n]
    test_data =  dataset[n:]
    
    
    return train_data,test_data
```

### Activation Function (Sigmoid)
Implementing **Sigmoid Activation function** for logistic regression prediction.

```python
import numpy as np
import math 
def sigmoid(row,parameters):
    hypothesis = parameters[0]
    for i in range(len(row)-1):
        hypothesis+=row[i]*parameters[i+1]
    return 1 / (1 + math.exp(-hypothesis))
```

### Loss Function
Implementing **Binary Cross Entropy** or **Log Loss Function** to check the closeness of prediction probability values from the actual/true values

```python
def loss_function_sigmoid(x,parameters):
    loss = 0
    for row in x:
        pred = sigmoid(row,parameters)
        y = row[-1]
        loss+= -(y*np.log(pred))+(-(1-y)*np.log(1-pred))
    mean_loss = loss/len(x)
    return mean_loss
```

### Optimization Algorithm Function
Implementing **Stochastic Gradient Descent (SGD)** function as an optimization algorithm to find model parameters that serves as best fit between predicted and actual values.

Function returns list of losses obtain from loss_function() and parameters.

```python
# Alpha value denotes the learning rate.
# Epochs denotes the number of complete passes of the dataset through the algorithm
# 'x' is the dataset passed as parameter in function
def sgd_sigmoid(x,epochs,alpha):
    
    parameters = [0]*len(x[0])
    loss_record = []
    n = len(x)
    
    for i in range(epochs):
        for row in x:
            pred = sigmoid(row,parameters)
            parameters[0] = parameters[0]-alpha*(pred-row[-1])
            for j in range(len(row)-1):
                parameters[j+1] = parameters[j+1]-alpha*(pred-row[-1])*row[j]
        loss_record.append(loss_function_sigmoid(x,parameters))
    return loss_record,parameters
```

### Accuracy Function
Implementing Accuracy checker to return accuracy (percentage value) for the model predicting actual/true value.

```python
def accuracy_func(pred,actual):
    counter = 0
    for i in range(len(actual)):
        if(pred[i]==actual[i]):
            counter+=1
    accuracy = (counter/len(actual))*100
    return accuracy
```

### Computing F-1 Score
Implementing set of functions to compute F-1 score for neural network model
- confusion_matrix_values function calculates and returns total number of true positives, true negatives, false negatives and false positives values based on a list of actual values from dataset and predicted values obtained.

```python
def confusion_matrix_values(y_act, y_pred):
    tp=0 # True Positive 
    tn=0 # True Negative
    fn=0 # False Negative
    fp=0 # False Positive
    
    for i in range(len(y_act)):
        #Increment True Positive variable if actual and predicted values are true
        if(y_act[i]==1 & y_pred[i]==1):
            tp+=1
            
        #Increment True Negative variable if actual and predicted values are false
        elif(y_act[i]==0 & y_pred[i]==0):
            tn+=1
            
        #Increment False Negative variable if actual values are true but predicted values are false
        elif(y_act[i]==1 & y_pred[i]==0):
            fn+=1
        
        #Increment False Positive variable if actual values are False but predicted values are True
        else:
            fp+=1
    return tp, tn, fp, fn
```

calc_precision function calculates **precision**

```python
def calc_precision(tp, fp):

    return (tp  * 100)/ float( tp + fp)
```

calc_recall function calculates **recall**

```python
def calc_recall(tp, fn):

    return (tp  * 100)/ float( tp + fn)
```

f1_Computation function calculates and returns **F1 Score**

```python
def f1_Computation(y_true, y_pred):
    # calculates the F1 score
    tp, tn, fp, fn = confusion_matrix_values(y_true, y_pred)
    precision = calc_precision(tp, fp)/100
    recall = calc_recall(tp, fn)/100
    f1_score = (2*precision*recall)/ (precision + recall)
    return f1_score
```

### Implementing the set of functions (Neural Network Model) on test data 
#### *using sigmoid activation function

```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt_sigmoid

def logistic_regression_unit_sigmoid(train_data,test_data):
    
   
    # Epochs denotes the number of complete passes of the dataset through the algorithm
    epochs = 1000
    
     # Alpha value denotes the learning rate.
    alpha = 0.001
    
    loss_record,parameters = sgd_sigmoid(train_data,epochs,alpha)
    
    #List of predicted values
    predictions = []
    
    for i in test_data:
        pred = sigmoid(i,parameters)
        predictions.append(round(pred))
        
    #List of actual values
    actual_values = [i[-1] for i in test_data]    
    accuracy = accuracy_func(predictions,actual_values)
    
    iterations = [i for i in range(1,epochs+1)]
    
    f1_score = f1_Computation(actual_values, predictions)
    
    plt.plot(iterations,loss_record)
    plt.show()
    return accuracy,f1_score
```

**Result:**

```python
accuracy,f1_score = logistic_regression_unit_sigmoid(train_data,test_data)
print("\033[1mUsing Sigmoid -\033[0m \nAccuracy:{} \nF1 Score for Logistic Regression (Sigmoid):{}".format(accuracy,f1_score))
```

### Activation Function (ReLU)
Implementing **Rectified Linear Unit (ReLU)** which will output the input directly if it is positive, otherwise, it will output zero.

```python
def relu(row,parameters):
    hypothesis = parameters[0]
    for i in range(len(row)-1):
        hypothesis+=row[i]*parameters[i+1]
    return max(0,hypothesis)
```

### Loss Function
Implementing **Binary Cross Entropy** or **Log Loss Function** to check the closeness of prediction probability values from the actual/true values

```python
def loss_function_relu(x,parameters):
    loss = 0
    for row in x:
        pred = relu(row,parameters)
        y = row[-1]
        loss+= -(y*np.log(pred))+(-(1-y)*np.log(1-pred))
    mean_loss = loss/len(x)
    return mean_loss
```

### Optimization Algorithm Function
Implementing Stochastic Gradient Descent (SGD) function as an optimization algorithm to find model parameters that serves as best fit between predicted and actual values.

Function returns list of losses obtain from loss_function() and parameters.

```python
# Alpha value denotes the learning rate.
# Epochs denotes the number of complete passes of the dataset through the algorithm
# 'x' is the dataset passed as parameter in function
def sgd_relu(x,epochs,alpha):
    
    parameters = [0]*len(x[0])
    
    #List of losses obtained from loss_function
    loss_record = []
    n = len(x)
    
    for i in range(epochs):
        for row in x:
            pred = relu(row,parameters)
            parameters[0] = parameters[0]-alpha*(pred-row[-1])
            for j in range(len(row)-1):
                parameters[j+1] = parameters[j+1]-alpha*(pred-row[-1])*row[j]
        loss_record.append(loss_function_relu(x,parameters))
    return loss_record,parameters
```

### Implementing the set of functions (Neural Network Model) on test data 
#### *using relu activation function

```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt_sigmoid

def logistic_regression_unit_relu(train_data,test_data):
    
    
    # Epochs denotes the number of complete passes of the dataset through the algorithm
    epochs = 1000
    
    # Alpha value denotes the learning rate.
    alpha = 0.001
    
    
    loss_record,parameters = sgd_relu(train_data,epochs,alpha)
    
    #List of predicted values
    predictions = []
    
    for i in test_data:
        pred = relu(i,parameters)
        predictions.append(round(pred))
    
    #List of actual values
    actual_values = [i[-1] for i in test_data]    
    accuracy = accuracy_func(predictions,actual_values)
    
    iterations = [i for i in range(1,epochs+1)]
    
    f1_score = f1_Computation(actual_values, predictions)
    
    plt.plot(iterations,loss_record)
    plt.show()
    return accuracy,f1_score
```

**Result:**

```python
accuracy,f1_score = logistic_regression_unit_relu(train_data,test_data)
print("\033[1mUsing ReLU -\033[0m \nAccuracy:{} \nF1 Score for Logistic Regression (ReLU):{}".format(accuracy,f1_score))
```
